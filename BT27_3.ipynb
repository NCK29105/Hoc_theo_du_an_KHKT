```python
def calculate_tax(income):
    if income < 5000:
        return 0
    elif income < 10000:
        return income * 0.1
    else:
        return income * 0.2


def test_calculate_tax():
    assert calculate_tax(3000) == 0

    assert calculate_tax(7000) == 7000 * 0.1

    assert calculate_tax(12000) == 12000 * 0.2

    print("Tất cả test đều PASS!")

if __name__ == "__main__":
    test_calculate_tax()
```

Bài 2

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


def test_is_prime():
    assert is_prime(1) == False
    assert is_prime(4) == False
    assert is_prime(18) == False

    assert is_prime(2) == True
    assert is_prime(3) == True
    assert is_prime(17) == True
    assert is_prime(19) == True

    print("Tất cả test đều PASS!")

if __name__ == "__main__":
    test_is_prime()
```

Bài 3

```python
class User:
    def __init__(self, username):
        self.username = username

    def is_admin(self):
        return self.username == "admin"

def test_user():
    u1 = User("admin")
    assert u1.is_admin() == True

    u2 = User("guest")
    assert u2.is_admin() == False

    u3 = User("user123")
    assert u3.is_admin() == False

    print("Tất cả test đều PASS!")


if __name__ == "__main__":
    test_user()
```

```python
from unittest.mock import patch

def send_welcome_email(email):
    print(f"Sending email to {email}")


def test_send_email():
    with patch("builtins.print") as mock_print:
        send_welcome_email("test@gmail.com")

        # Kiểm tra print được gọi đúng nội dung
        mock_print.assert_called_with("Sending email to test@gmail.com")

    print("Test PASS!")


if __name__ == "__main__":
    test_send_email()
```

```python
import requests
from unittest.mock import patch, Mock

def fetch_user():
    response = requests.get("https://api.example.com/user")
    return response.json()


def test_fetch_user():
    mock_data = {"name": "Khanh", "age": 20}

    mock_response = Mock()
    mock_response.json.return_value = mock_data

    with patch("requests.get", return_value=mock_response) as mock_get:
        result = fetch_user()

        mock_get.assert_called_with("https://api.example.com/user")

        assert result == mock_data

    print("Test PASS!")


if __name__ == "__main__":
    test_fetch_user()
```

```python
from datetime import datetime
from unittest.mock import patch

def is_weekend():
    today = datetime.now().weekday()
    return today >= 5


def test_is_weekend():
    with patch(__name__ + ".datetime") as mock_datetime:
        mock_datetime.now.return_value.weekday.return_value = 4
        assert is_weekend() == False

    with patch(__name__ + ".datetime") as mock_datetime:
        mock_datetime.now.return_value.weekday.return_value = 5
        assert is_weekend() == True

    print("Test PASS!")

if __name__ == "__main__":
    test_is_weekend()
```

Bài 7

```python
def is_strong(password):
    return len(password) >= 8 and any(c.isdigit() for c in password)


def test_is_strong():
    assert is_strong("abc") == False              
    assert is_strong("abcdefgh") == False         
    assert is_strong("1234567") == False          

    assert is_strong("abc12345") == True
    assert is_strong("password1") == True
    assert is_strong("A1b2c3d4") == True

    print("Tất cả test đều PASS!")

if __name__ == "__main__":
    test_is_strong()
```

```python
def clean_input(s):
    return s.strip().lower().replace(" ", "_")


def test_clean_input():
    assert clean_input("  Hello World  ") == "hello_world"

    assert clean_input("PYTHON") == "python"

    assert clean_input("Hello   World") == "hello___world"

    assert clean_input("test case") == "test_case"

    print("Tất cả test đều PASS!")


if __name__ == "__main__":
    test_clean_input()
```

```python
def safe_divide(a, b):
    if b == 0:
        return None
    return a / b

def test_safe_divide():
    assert safe_divide(10, 2) == 5
    assert safe_divide(5, 5) == 1

    assert safe_divide(-4, 2) == -2

    assert safe_divide(10, 0) is None

    print("Tất cả test đều PASS!")

if __name__ == "__main__":
    test_safe_divide()
```
